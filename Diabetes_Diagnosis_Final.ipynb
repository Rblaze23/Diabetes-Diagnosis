```python
#Ramy Lazghab Work :)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve, f1_score
from imblearn.over_sampling import SMOTE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import shap
import lime
import lime.lime_tabular
```

## 📂 1. Dataset Description

The dataset used in this project relates to **diabetes diagnosis**, containing several key medical and lifestyle features:

- `Glucose`: blood sugar level
- `BloodPressure`: diastolic blood pressure
- `InsulinLevel`: insulin concentration
- `BMI`: body mass index
- `Age`: patient’s age
- `Lifestyle`: categorical description of activity level
- `Diabetes`: target variable (1: diabetic, 0: non-diabetic)

The goal is to build predictive models to **accurately identify diabetic patients** based on these features.

To begin, I explored the dataset using:
- `.head()`, `.info()`: for structure, data types, missing values
- `.describe()`: for basic statistical distributions
- Histograms and correlation matrix: for visual understanding

🔍 **Why this step?**  
Understanding the **structure and quality of the data** is essential before any transformation or modeling. It helps identify inconsistencies, biases, or patterns that might guide the next steps.

```python


# Chargement des données
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Data mining/diabetes_diagnosis.xlsx')

# Contexte
print("=== Contexte du Projet ===")
print("""
Objectif : Prédire le diabète (1) ou non (0) basé sur :
- Glucose : Niveau de glucose sanguin
- BloodPressure : Pression artérielle
- InsulinLevel : Niveau d'insuline
- BMI : Indice de masse corporelle
- Age : Âge du patient
- Lifestyle : Mode de vie (Sedentary/Moderate/Active)
""")

# Exploration
print("\n=== Exploration Initiale ===")
print(f"Lignes: {df.shape[0]}, Colonnes: {df.shape[1]}")
print("\nAperçu:")
print(df.head())

print("\nTypes de données:")
print(df.dtypes)

print("\nValeurs manquantes:")
print(df.isnull().sum())

print("\nStatistiques descriptives:")
print(df.describe())

# Visualisations
plt.figure(figsize=(8,6))
sns.countplot(x='Diabetes', data=df)
plt.title('Distribution des diagnostics')
plt.show()

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Matrice de corrélation')
plt.show()

num_cols = ['Glucose', 'BloodPressure', 'InsulinLevel', 'BMI', 'Age']
df[num_cols].hist(figsize=(12,10), bins=20)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='Lifestyle', hue='Diabetes', data=df)
plt.title('Diabète par mode de vie')
plt.show()
```

## 🧹 2. Data Preprocessing

This step included multiple cleaning and preparation tasks:

1. **Invalid values correction**:
   - Replaced medically implausible values (e.g., negative `BMI` or `BloodPressure`)
2. **Missing values imputation**:
   - Numerical features: median (to reduce impact of outliers)
   - Categorical: most frequent value
3. **Label Encoding** for the categorical feature `Lifestyle`
4. **Feature scaling** using `StandardScaler` on continuous variables
5. **Class imbalance correction** using **SMOTE** (Synthetic Minority Oversampling Technique)

🧠 **Why these choices?**  
- Imputation avoids data loss from dropping rows and maintains dataset size.
- Standardization is crucial for models sensitive to scale (like logistic regression and neural networks).
- SMOTE was chosen because the dataset had **imbalanced target classes** (more non-diabetics), which could bias model predictions toward the majority class.


```python
# Nettoyage des valeurs négatives
for col in ['BloodPressure', 'BMI']:
    df[col] = df[col].apply(lambda x: np.nan if x < 0 else x)

# Imputation
num_imputer = SimpleImputer(strategy='median')
df[num_cols] = num_imputer.fit_transform(df[num_cols])

cat_imputer = SimpleImputer(strategy='most_frequent')
df[['Lifestyle']] = cat_imputer.fit_transform(df[['Lifestyle']])

# Encodage
le = LabelEncoder()
df['Lifestyle'] = le.fit_transform(df['Lifestyle'])
print("\nEncodage Lifestyle:")
print(df['Lifestyle'].unique())

# Normalisation
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Gestion du déséquilibre
X = df.drop('Diabetes', axis=1)
y = df['Diabetes']
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Split train-test
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res,
    test_size=0.3,
    random_state=42
)
print(f"\nTrain: {X_train.shape}, Test: {X_test.shape}")
```

## 🧠 3. Transparent Models

To begin modeling, I applied two interpretable models:

- **Logistic Regression**: offers direct interpretability through feature coefficients.
- **Decision Tree Classifier**: explains predictions in a step-by-step flowchart manner.

Each model was evaluated using:
- Accuracy, Precision, Recall, F1-score
- Confusion matrix to understand classification behavior

🧠 **Why start with transparent models?**  
In medical applications, interpretability is vital. Doctors and stakeholders often need **justifications** behind predictions. These models are easy to explain, debug, and validate, and they provide a **baseline** for comparison with more complex models later.


```python
# Régression Logistique
lr = LogisticRegression(class_weight='balanced', max_iter=1000)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print("=== Régression Logistique ===")
print(classification_report(y_test, y_pred_lr))
print("Matrice de confusion:")
print(confusion_matrix(y_test, y_pred_lr))

# Importance des caractéristiques
coeff_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': lr.coef_[0]
}).sort_values('Coefficient', ascending=False)
print("\nImportance des variables:")
print(coeff_df)

# Arbre de Décision
dt = DecisionTreeClassifier(class_weight='balanced', max_depth=3, random_state=42)
dt.fit(X_train, y_train)

plt.figure(figsize=(20,10))
plot_tree(dt, feature_names=X.columns, class_names=['Non-Diabétique','Diabétique'], filled=True)
plt.title("Arbre de décision")
plt.show()

print("\nImportance des variables:")
print(pd.DataFrame({
    'Feature': X.columns,
    'Importance': dt.feature_importances_
}).sort_values('Importance', ascending=False))
```

## ⚙️ 4. Black-Box Models + Interpretability

To improve prediction performance, I applied two ensemble models:

- **Random Forest**: multiple decision trees combined for robustness
- **XGBoost**: gradient boosting algorithm known for its efficiency and accuracy

As these are black-box models, I used interpretability tools:

- **SHAP (SHapley Additive exPlanations)**:
  - To understand the **global and local feature impact**
  - Based on game theory, fair and consistent
- **LIME (Local Interpretable Model-Agnostic Explanations)**:
  - To interpret **individual predictions** via local surrogate models

🧠 **Why use black-box models?**  
They often capture **complex, non-linear relationships** and provide better performance. However, to maintain **trust and transparency**, especially in healthcare, tools like SHAP and LIME are crucial to explain the predictions and identify key risk factors.


```python
### Random Forest
rf = RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

### XGBoost
xgb = XGBClassifier(
    scale_pos_weight=sum(y_train == 0) / sum(y_train == 1),
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)
xgb.fit(X_train, y_train)

# Interprétation SHAP (Random Forest)


explainer_rf = shap.TreeExplainer(rf)
shap_values_rf = explainer_rf(X_test)  # Nouvelle API

print(f"Shape SHAP RF : {shap_values_rf.shape}")
print(f"X_test shape : {X_test.shape}")

# Résumé Global
plt.figure(figsize=(12, 8))
print(f"Shape SHAP RF : {shap_values_rf.shape}")
if len(shap_values_rf.shape) == 3:  # Multiclass
    shap.summary_plot(shap_values_rf[..., 1], X_test, feature_names=X.columns.tolist())
else:
    shap.summary_plot(shap_values_rf, X_test, feature_names=X.columns.tolist())


# Interprétation SHAP (XGBoost)


explainer_xgb = shap.TreeExplainer(xgb)
shap_values_xgb = explainer_xgb(X_test)

plt.figure(figsize=(12, 8))
print(f"Shape SHAP XGB : {shap_values_xgb.shape}")
print(f"X_test shape : {X_test.shape}")
if len(shap_values_xgb.shape) == 3:
    shap.summary_plot(shap_values_xgb[..., 1], X_test, feature_names=X.columns.tolist())
else:
    shap.summary_plot(shap_values_xgb, X_test, feature_names=X.columns.tolist())


# Interprétation LIME (sur RF, 1 individu)


explainer_lime = lime.lime_tabular.LimeTabularExplainer(
    X_train.values,
    feature_names=X.columns,
    class_names=['Non-Diabétique', 'Diabétique'],
    mode='classification'
)

# Exemple sur le premier individu
i = 0
exp = explainer_lime.explain_instance(
    X_test.iloc[i].values,
    rf.predict_proba,
    num_features=5
)
exp.show_in_notebook()
```

## 🔧 5. Model Optimization

To improve model performance, I tuned hyperparameters for the Random Forest using **GridSearchCV**.

Parameters tested:
- `n_estimators`: number of trees
- `max_depth`: maximum depth of each tree
- `min_samples_split`: minimum samples to split a node
- `class_weight`: used 'balanced' to handle class imbalance

I used **5-fold cross-validation** to ensure the results were robust and not overfitting.

🧠 **Why this method?**  
Grid Search systematically tests combinations of parameters. Cross-validation ensures that the model generalizes well and does not just memorize the training data. Optimizing a model this way maximizes its **predictive power** while minimizing variance.


```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'class_weight': ['balanced', None]
}

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)
grid_search.fit(X_train, y_train)

best_rf = grid_search.best_estimator_
print(f"\nMeilleurs paramètres: {grid_search.best_params_}")
print(f"Meilleur score AUC-ROC: {grid_search.best_score_:.2f}")
```

## 🧬 6. Neural Network

I built a **feedforward neural network (MLP)** using Keras with the following architecture:

- Dense layers with ReLU activation
- Dropout layers to reduce overfitting
- Final sigmoid activation for binary classification
- Compiled with binary cross-entropy and `Adam` optimizer

Training included **early stopping** to prevent overfitting and save the best model.

📊 I plotted loss and accuracy curves to monitor learning progress.

🧠 **Why neural networks?**  
They are powerful for discovering **non-linear patterns** in data, especially in complex domains like medicine. Although they are less interpretable, they often improve prediction metrics when trained carefully.


```python
# Architecture
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Poids des classes
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights = dict(enumerate(class_weights))

# Entraînement
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    class_weight=class_weights,
    callbacks=[EarlyStopping(patience=5)],
    verbose=1
)
#visualisation loss and accuracy
def plot_history(history, metric=None):
  """plot loss and passed metric.  metric is passed as string and must match 'metric'
  argument in the compile step"""
  fig, axes = plt.subplots(2,1, figsize = (5,10))
  axes[0].plot(history.history['loss'], label = "train")
  axes[0].plot(history.history['val_loss'], label='test')
  axes[0].set_title('Loss')
  axes[0].legend()
  if metric:
    axes[1].plot(history.history[metric], label = 'train')
    axes[1].plot(history.history['val_' + metric], label = 'test')
    axes[1].set_title(metric)
    axes[1].legend()
  plt.show()
# Courbes d'apprentissage
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Courbe d\'apprentissage')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

## 📈 7. Final Evaluation & Conclusion

I compared all models based on:

- Accuracy
- Precision, Recall, F1-score
- ROC AUC
- Precision-Recall curves (especially useful in imbalanced datasets)

📊 A summary table highlights the performance of each model.

✅ **Final Thoughts**:
- **Optimized Random Forest** provided the best trade-off between accuracy and interpretability (with SHAP explanations).
- **Neural Network** performed well but remains a black-box.
- **Logistic Regression** is highly interpretable and useful as a baseline.

💡 This project demonstrates the value of combining **predictive performance** with **interpretability tools** in healthcare, ensuring both accuracy and ethical transparency in decision-making.


```python
# Courbe Precision-Recall (Random Forest)

# Prédictions probabilistes
probas_rf = best_rf.predict_proba(X_test)[:, 1]

# Calcul des points Precision-Recall
precision, recall, _ = precision_recall_curve(y_test, probas_rf)

# Affichage de la courbe
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label='Random Forest')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Courbe Precision-Recall (Random Forest)')
plt.grid(True)
plt.legend()
plt.show()

# Comparaison
results = []

# Liste
models_to_compare = [
    ('Régression Logistique', lr),
    ('Arbre de Décision', dt),
    ('Random Forest', best_rf),
    ('XGBoost', xgb),
    ('Réseau de Neurones', model)  # keras
]

for name, mdl in models_to_compare:
    # Gestion du cas particulier du réseau de neurones
    if name == 'Réseau de Neurones':
        y_pred = (mdl.predict(X_test) > 0.5).astype(int)
    else:
        y_pred = mdl.predict(X_test)

    # Enregistrement des métriques
    results.append({
        'Modèle': name,
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'Accuracy': accuracy_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
        'AUC-ROC': roc_auc_score(y_test, y_pred)
    })

# Affichage des résultats triés par AUC-ROC
df_results = pd.DataFrame(results).sort_values('AUC-ROC', ascending=False)
display(df_results)
```

```python
import joblib

# save
joblib.dump(best_rf, 'modele_diabete.pkl')
joblib.dump(scaler, 'scaler.pkl')

# Fonction de prédiction
def predire_diabete(glucose, tension, insuline, imc, age, mode_vie):
    """
    Mode vie:
    0 = Actif, 1 = Modéré, 2 = Sédentaire
    Retourne la probabilité de diabète (0-1)
    """
    data = pd.DataFrame([[
        glucose, tension, insuline, imc, age, mode_vie
    ]], columns=X.columns)

    data[num_cols] = scaler.transform(data[num_cols])
    return best_rf.predict_proba(data)[0,1]

# Exemple
print("Probabilité:", predire_diabete(120, 80, 60, 30, 45, 2))
```
